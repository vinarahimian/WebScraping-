### The goal here is to re-create recursive links of a website and to extract certain information from them


We're going to a website which gives information on all restaurants of any city in Portugal. The information includes the address, phone number and website of restaurants in any specific city. Here's we're writing an alghorithm that scrapes information on all restaurants in the beautiful city of Porto in Portugal

```python
import requests
import pprint

#the website we want to start scraping info
demo_url = 'https://www.portugalio.com/restaurantes/porto/'
r = requests.get(demo_url)
```

```python
#it means ok we have a connection here (when it's 200)
r.status_code
```

```python
#First post Xpath: //*[@id="list-container"]/div[1]/div[1]/div/div[1]/div[2]/p[1]/a
#Last post Xpath: //*[@id="list-container"]/div[1]/div[17]/div/div[1]/div[2]/p[1]/a

#now here we create the Xpath for all of the postings in the website
xpaths=[]
for n in range (1,18):
    first= '//*[@id="list-container"]/div[1]/div['
    scnd = ']/div/div[1]/div[2]/p[1]/a/@href' #@href make the code to go into each link and get the data
    path = first+str(n)+scnd
    xpaths.append(path)

pprint.pprint(xpaths)
```

```python
from lxml import html
```

```python
tree = html.fromstring(r.text) #A tree is a structured html (understands that it's not plain text and has data in it)
```

```python
posts_url=[]
to_scrape=[] # <---- single page url to scrape
#categories=['restaurantes','bar']
#cities = ['porto', 'lisboa']
mydomain = 'http://www.portugalio.com'

for p in xpaths:
    #print(p)
    result = tree.xpath(p) #the result is going to be the content of the path (here is the name of the restaurants)
    #print(result)
    
    if result: #to get rid of the empty bracket (which was advertisement in the real website)
        posts_url.append(mydomain+result[0]) #we put [0] to not have [[]] and have just a single list + also we're adding the domain name to the name of restaurant to get the name of the webpages
    
pprint.pprint(posts_url)
```

```python
#Here we want to know how many posts we have per page
results_path='//*[@id="list-filter"]/div[5]/div/span[2]/text()' #the total number of restaurants in the website
results_number = tree.xpath(results_path)
print(results_number)
number = results_number[0].split(' ')[0] #the total number
print(number)
```

```python
import math
res_per_page = 15
if results_number:
    pages= math.ceil(int(number)/res_per_page)
    print(pages)
```

```python
def makeUrl(pnumber):
    d=[]
    for p in range (1,pnumber+1):
        url = mydomain+'/restaurantes/porto/'+str(p)+'.html'
        d.append(url)
        print(url)
    return d
        
if pages<20:
    to_scrape=makeUrl(pages)
    print('less than 20')
else: 
    to_scrape = makeUrl(20)
    print('more than 20')
```

```python
pprint.pprint(to_scrape)
```

```python
paths = {'address': '//*[@id="page-title-container"]/div/text()',
         'pnumber': '//*[@id="second-container"]/div[2]/div/div/div/div[3]/a/text()',
         'website':'//*[@id="second-container"]/div[4]/div[1]/div/div[3]/a[1]/@href'
        } 
```

```python
def justbadfunction():
    bigcontainer=[]
    #if we have some links inside the array 
    if to_scrape:
        print('we have links to scrape')
        
        for s in to_scrape:
            #is opening the url to scrape(multiple page list)
            r = requests.get(s)
            #we axtract the tree from the simple text
            tree = html.fromstring(r.text)
            
            #now every time we open one of these wonderful multiple page we want to extract a single post url and then
            #open the url
            for p in xpaths:
                #here is extracting link by link(single post)
                link = tree.xpath(p)
                #now is opening the first one...
                if link:
                    r = requests.get(mydomain+link[0])
                    #extracting the tree
                    branch = html.fromstring(r.text)
                    #an empty dictionary is waiting for data
                    d={}
                    for key, value in paths.items():
                        #i want to update my little empty dictionary
                        d.update({key: branch.xpath(value)})
                    
                    #putting a single result page values inside our bigcontainer    
                    bigcontainer.append(d)    
                
                    pprint.pprint (d)
            pprint.pprint(s)
    else:
        print ('we are bored')
        

justbadfunction()
        




```
